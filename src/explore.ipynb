```python
df['COUNTY_NAME'] = pd.factorize(df['COUNTY_NAME'])[0]
df['STATE_NAME'] = pd.factorize(df['STATE_NAME'])[0]

df.info()
```

```python
x = df.drop('Obesity_number', axis=1)
y = df['Obesity_number']
```

```python
correlaciones = x.corrwith(y)

print(correlaciones)
```

```python
altas_correlaciones = correlaciones[correlaciones>0.95]
altas_correlaciones
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( x , y, test_size=0.2, random_state=42)


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_values = scaler.fit_transform(X_train)
x_test_values = scaler.transform(X_test) 

x_train_esc = pd.DataFrame(x_train_values, index=X_train.index, columns=X_train.columns)
x_test_esc = pd.DataFrame(x_test_values, index=X_test.index, columns=X_test.columns)
```

```python
from sklearn.linear_model import LinearRegression , Lasso

modelRL = LinearRegression()
modelLasso = Lasso()

modelRL.fit(x_train_esc,y_train)
modelLasso.fit(x_train_esc,y_train)
```

```python
y_pred_rl = modelRL.predict(x_test_esc)
y_pred_lasso = modelLasso.predict(x_test_esc)
```

```python
from sklearn.metrics import root_mean_squared_error , r2_score

rmse_rl = root_mean_squared_error(y_test, y_pred_rl)
r2_rl = r2_score(y_test, y_pred_rl)
rmse_lasso = root_mean_squared_error( y_test , y_pred_lasso)
r2_lasso = r2_score(y_test , y_pred_lasso)

print(f'Para la regresiÃ³n lineal se tuvo un RMSE de: {rmse_rl} y un R2 de: {r2_rl}')
print(f'Para Lasso se tuvo un RMSE de: {rmse_lasso} y un R2 de: {r2_lasso}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

espacio = np.linspace(0,20,40)
espacio
```

```python
R2 = []
for i in espacio:
  
    model_Lasso_Variable = Lasso(alpha = i, max_iter=100000)
    model_Lasso_Variable.fit(x_train_esc, y_train)
    
    y_pred_lasso_actual = model_Lasso_Variable.predict(x_test_esc)
    
   
    score = r2_score(y_test, y_pred_lasso_actual)
    R2.append(score)

print(R2)

```

```python
import matplotlib.pyplot as plt

plt.plot(espacio, R2)
plt.show()
```
